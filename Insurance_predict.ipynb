```python
medical_df=pd.read_csv('expenses.csv')
```

```python
def rmse(targets,predictions):
  return np.sqrt(np.mean(np.square(targets-predictions)))
```

```python
smoker_code={'yes':1,'no':0}
medical_df['smoker_code']=medical_df['smoker'].map(smoker_code)
```

```python
sex_code={'male':1,'female':0}
medical_df['sex_code']=medical_df['sex'].map(sex_code)
```

```python
encode=preprocessing.OneHotEncoder()
encode.fit(medical_df[['region']])
encode.categories_
one_hot=encode.transform(medical_df[['region']]).toarray()
one_hot
medical_df[['northeast','northwest','southeast','southwest']]=one_hot
```

```python
input,targets= medical_df[['age','bmi','children','smoker_code','sex_code','northeast','northwest','southeast','southwest']],medical_df['charges']
model=LinearRegression().fit(input,targets)
predictions=model.predict(input)
loss=rmse(targets,predictions)
print('Loss',loss)
```

```python
input,targets= medical_df[['age','bmi','children','smoker_code','sex_code','northeast','northwest','southeast','southwest']],medical_df['charges']
model_rf=RandomForestRegressor().fit(input,targets)
predictions=model_rf.predict(input)
loss=rmse(targets,predictions)
print('Loss',loss)
```
